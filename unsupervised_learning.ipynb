```python
# import libraries
import pandas as pd # type: ignore
import matplotlib.pyplot as plt # type: ignore
import seaborn as sns # type: ignore
from sklearn.cluster import KMeans # type: ignore
```

```python
# load dataset
df = pd.read_csv("customer_data_records.csv")
```

```python
#select feature for clustering
features = df[["Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
#Aplly K-Means clustering
# I will use only two clusters
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(features)
```

```python
# Add cluster labels to the original dataframe(dataset)
df["Cluster"] = kmeans.labels_
print(df)
```

```python
# Visualize the clusters
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Annual Income (k$)", y="Spending Score (1-100)", hue="Cluster")
plt.title("Customer Cluster Based on Income and Spending")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
#plt.legend(title="Cluster")
plt.show()
```
